```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
```

```python
root = 'carpark datasets'
```

```python
def clean_cd(data):
  data.sort_values(['number', 'time'], inplace=True)
#   data = data.assign(usage=lambda x: (x.total-x.available)/x.total)
  nd = pd.DataFrame(data.number.value_counts())
  nd = nd[nd.number<=288]
  data = data[data.number.isin(nd.index)]
  data.time = pd.to_datetime(data.time)
  data.time = data.time.dt.hour
  data = data.groupby(['number', 'time'],as_index=False).mean()
  return data.reset_index(drop=True).assign(count=1)
```

```python
jeff = pd.read_csv(f'{root}/output_2020/2020-03-22.csv')
jeff = clean_cd(jeff)
jeff
```

```python
import glob
def mean_year(year):
    dir = f'carpark datasets/output_'+str(year)
    all_files = glob.glob(dir + "/*.csv")
    frick = pd.DataFrame()
    for f in all_files:
        jeff = clean_cd(pd.read_csv(f))
        frick = frick.append(jeff).groupby(['number', 'time'],as_index=False).sum()
        print(f)

    frick.total /= frick['count']
    frick.available /=frick['count']
    frick.reset_index(drop=True).to_csv(str(year)+'.csv', index=False)
```

```python
for i in ['2020', '2019', '2018']:
    mean_year(i)
```
